```python
# Question 4: What is a binomial distribution, and how is it used in probability?
```

```python
"""
- A binomial distribution gives the probability of getting exactly k successes in n 

independent Bernoulli trials , where each trial has a fixed success probability p

- CONDITIONS FOR BINOMIAL DISTRIBUTION

1. Fixed number of trials n 

2. Each trial has two outcomes(success/failure)

3. Probability of Success p is the same for each trial

4. All trials are independent
"""
```

```python
from scipy.stats import binom
import matplotlib.pyplot as plt

# parameters
n=10  # number of trials
p= 0.5 # probability of success

# probability of exactly k success

k=4
prob= binom.pmf(k,n,p)
print(f"P(X={k}) = " , prob)

# plot entire distribution
x= list(range(n+1))
pmf = binom.pmf(x, n, p)

plt.stem(x, pmf , basefmt=" ")
plt.title(f"Binomial distribution (n={n} , p={p})")
plt.xlabel("Number of Successes")
plt.ylabel("Probaility")
plt.grid(True)
plt.show()
```

```python
# What is the Central Limit Theorem (CLT), and why is it critical in statistics?
```

```python
"""
The Central Limit Theorem states that if we take a large number of independent samples

from a population with a finite mean and variance, the distribution of the sample mean

tends to become normal, even if the original population is not normally distributed. 

It is critical in statistics because it allows us to apply normal distribution–based

methods such as confidence intervals and hypothesis testing in many real-world 

situations, making statistical analysis practical and reliable.


"""
```

```python
# Question 7: What is the significance of confidence intervals in statistical analysis?
```

```python
"""
Confidence intervals are used to estimate an unknown population parameter, such as 

the mean or proportion, by providing a range of values rather than a single point

estimate. This range is constructed so that it is likely to contain the true population 

parameter with a specified level of confidence, such as 95%. The significance of

confidence intervals lies in the fact that they quantify the uncertainty in sample-based

estimates and give a clearer, more reliable understanding of how accurate an estimate is.


"""
```

```python
# Question 8: What is the concept of expected value in a probability distribution?
```

```python
"""
The expected value is the average or mean value of a random variable over a large

number of trials. It represents the long-run outcome of a random experiment by weighting

each possible value of the random variable by its probability. In a probability 

distribution, the expected value provides a single numerical summary that describes

the center of the distribution and helps in understanding what value is most likely

to occur on average.

"""
```

```python
#Question 9: Write a Python program to generate 1000 random numbers from a normal
#distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
#deviation using NumPy, and draw a histogram to visualize the distribution.
#(Include your Python code and output in the code box below.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random numbers from a normal distribution
data = np.random.normal(loc=50, scale=5, size=1000)

# Compute mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

print("Mean:", mean)
print("Standard Deviation:", std_dev)

# Draw histogram
plt.hist(data, bins=30)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Normal Distribution (Mean=50, Std=5)")
plt.show()
```

```python
"""
Question 10: You are working as a data analyst for a retail company. The company has

collected daily sales data for 2 years and wants you to identify the overall sales trend.

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,

 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
 
● Explain how you would apply the Central Limit Theorem to estimate the average sales

with a 95% confidence interval.

● Write the Python code to compute the mean sales and its confidence interval.

(Include your Python code and output in the code box below.)
"""
```

```python
import numpy as np
from scipy import stats

# Given daily sales data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Convert to NumPy array
data = np.array(daily_sales)

# Calculate mean and standard deviation
mean_sales = np.mean(data)
std_sales = np.std(data, ddof=1)  # sample standard deviation
n = len(data)

# Standard error
std_error = std_sales / np.sqrt(n)

# 95% confidence interval
confidence_interval = stats.norm.interval(0.95, loc=mean_sales, scale=std_error)

print("Mean Daily Sales:", mean_sales)
print("95% Confidence Interval:", confidence_interval)
```
