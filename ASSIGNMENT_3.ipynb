```python
#Question 1 : What are Type I and Type II errors in hypothesis testing, and how do they
#impact decision-making?
```

```python
"""

- Type I and Type II errors** are mistakes that can occur in hypothesis testing and affect decision-making.

- A Type I error happens when we reject the null hypothesis even though it is true. It is also called a false positive.

This can lead to wrong actions, such as assuming an effect exists when it actually doesn’t.

- A Type II error happens when we fail to reject the null hypothesis even though it is false. It is also called a false negative.

This can cause missed opportunities, such as failing to detect a real effect.

In practice, reducing one type of error often increases the other, so decision-making depends on which error is more costly in a given situation.


"""
```

```python
# Question 2:What is the P-value in hypothesis testing, and how should it be interpreted
# in the context of the null hypothesis?
```

```python
"""


- The p-value in hypothesis testing tells us how likely it is to observe the given data, or something more extreme, assuming the null hypothesis 

is true.

- If the p-value is very small, it means the observed result is unlikely under the null hypothesis, so we reject the null hypothesis. If the

p-value is large, the result is consistent with the null hypothesis, so we do not reject it.

- Usually, the p-value is compared with a significance level like 0.05. When the p-value is less than or equal to this value,

we reject the null hypothesis; otherwise, we fail to reject it. In simple terms, the p-value shows how strong the evidence is

against the null hypothesis.

"""
```

```python
# Question 3:Explain the difference between a Z-test and a T-test, including when to use  each.
```

```python
"""
- A Z-test and a T-test are both used to compare sample data with a population or between groups, but they are used in different situations.

- A Z-test is used when the population standard deviation is known and the sample size is large, usually 30 or more. It assumes 

the data follows a normal distribution. Because the population variance is known, the Z distribution is used.

- A T-test is used when the population standard deviation is unknown and the sample size is small. It is more commonly used in

real-world problems. The T distribution accounts for extra uncertainty by having heavier tails.

- In short, use a Z-test for large samples with known population variance, and use a T-test for small

samples when the population variance is unknown.


"""
```

```python
# Question 4:What is a confidence interval, and how does the margin of error influence its width and interpretation?
```

```python
"""
- A confidence interval is a range of values used to estimate an unknown population parameter, such as a mean. It gives an idea 

of where the true value is likely to lie based on sample data.

- The margin of error determines how wide the confidence interval is. A larger margin of error makes the interval wider, meaning less precision,

while a smaller margin of error makes it narrower, meaning more precision.

- In interpretation, a wider confidence interval shows more uncertainty about the estimate, whereas a narrower confidence interval indicates

greater confidence and accuracy in the estimated value.


"""
```

```python
# Question 5: Describe the purpose and assumptions of an ANOVA test. How does it extend hypothesis testing to more than two groups?
```

```python
"""
- The purpose of an ANOVA test is to check whether there are significant differences between the means of three or more groups. Instead of 

comparing groups one by one, ANOVA tests them together in a single analysis.

- ANOVA assumes that the data in each group is normally distributed, the variances of the groups are approximately equal, and the observations 

are independent of each other.

- It extends hypothesis testing to more than two groups by comparing the variation between group means with the variation within the groups. If

the between-group variation is much larger, ANOVA concludes that at least one group mean is different.


"""
```

```python
# Question 6: Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset.
```

```python
import numpy as np
from scipy import stats

# Given data
data = np.array([52, 55, 48, 50, 53, 49, 54, 51, 52, 50])
population_mean = 50
population_std = 10

# Sample statistics
sample_mean = np.mean(data)
n = len(data)

# Z-test calculation
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print("Sample Mean:", sample_mean)
print("Z-score:", z_score)
print("P-value:", p_value)
```

```python
# Question 8: Generate multiple samples from a non-normal distribution and implement the Central Limit Theorem using Python
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
population_size = 100000
sample_size = 30
num_samples = 1000

# Generate non-normal population (exponential distribution)
population = np.random.exponential(scale=2, size=population_size)

# Take multiple samples and compute means
sample_means = []
for _ in range(num_samples):
    sample = np.random.choice(population, sample_size)
    sample_means.append(np.mean(sample))

# Plot histogram of sample means
plt.hist(sample_means, bins=30, edgecolor='black')
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.title("Central Limit Theorem Demonstration")
plt.show()
```

```python
# Question 9: Write a Python function to calculate and visualize the confidence interval for a sample mean.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def confidence_interval_plot(data, confidence=0.95):
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    
    # Standard error
    se = std / np.sqrt(n)
    
    # Critical value (t-distribution)
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)
    
    margin_error = t_crit * se
    lower = mean - margin_error
    upper = mean + margin_error
    
    # Print results
    print("Sample Mean:", mean)
    print("Confidence Interval:", (lower, upper))
    
    # Visualization
    plt.errorbar(mean, 0, xerr=margin_error, fmt='o')
    plt.axvline(mean, linestyle='--')
    plt.title(f"{int(confidence*100)}% Confidence Interval for Mean")
    plt.xlabel("Value")
    plt.yticks([])
    plt.show()

# Example usage
sample_data = [45, 50, 52, 48, 49, 51, 47, 53, 50, 49]
confidence_interval_plot(sample_data)
```

```python
# Question 10: Perform a Chi-square goodness-of-fit test using Python to compare observed and expected distributions, and explain the outcome.
```

```python
import numpy as np
from scipy.stats import chisquare

# Observed and expected frequencies
observed = np.array([8, 9, 10, 11, 12, 10])
expected = np.array([10, 10, 10, 10, 10, 10])

# Chi-square test
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square statistic:", chi_stat)
print("P-value:", p_value)
```
