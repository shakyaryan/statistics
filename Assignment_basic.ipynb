Question 3: Define mean, median, and mode. Explain why these measures of central
tendency are important.

Ans->


- Mean

The mean is the average of a set of values. It is calculated by adding all the values and dividing by the total number of values.


[
\text{Mean} = \frac{\text{Sum of all values}}{\text{Number of values}}
]

- Median
  
The median is the middle value when the data is arranged in ascending or descending order. If the number of observations is even, the median is the average of the two middle values.

- Mode
The mode is the value that occurs most frequently in the data set. A data set can have one mode, more than one mode, or no mode.



- Importance of Measures of Central Tendency

        - Summarize data


                  - Mean, median, and mode give a single value that represents the entire data set, making large data easier to understand.

  
        - Understand data distribution

               - Mean shows the overall average
               - Median shows the central position (especially useful with outliers)
               - Mode shows the most common value

       - Comparison of datasets

  
                - These measures help compare different groups, such as average marks of two classes or income levels of two regions.

  
      -Decision making
   Used in real-life and data science applications like salary analysis, exam results, business performance, and customer behavior.



e.g

Marks: 10, 20, 20, 30, 100

* Mean = 36 (affected by outlier 100)
* Median = 20 (better representation)
* Mode = 20 (most common value)



Question 4: Explain skewness and kurtosis. What does a positive skew imply about the
data?


Ans- 

- Skewness

  Skewness is a statistical measure that describes the asymmetry of a data distribution around its central value. If the data is perfectly symmetrical, the skewness is zero. When the distribution is not symmetrical, it can be either positively skewed or negatively skewed. In a positively skewed (right-skewed) distribution, the right tail is longer than the left tail. In a negatively skewed (left-skewed) distribution, the left tail is longer than the right tail.

- Krutosis

Kurtosis is a statistical measure that describes the shape of a distribution in terms of the heaviness of its tails and the sharpness of its peak compared to a normal distribution. A distribution with normal kurtosis is called mesokurtic. A leptokurtic distribution has a sharper peak and heavier tails, indicating more extreme values or outliers. A platykurtic distribution has a flatter peak and lighter tails, indicating fewer extreme values.

- A positive skew implies that most of the data values are concentrated on the lower end of the distribution, while a few unusually large values extend the tail to the right. Because of these high values, the distribution is right-skewed and the mean is pulled toward the higher side, making the mean greater than the median, and the median greater than the mode.


Question 5: Implement a Python program to compute the mean, median, and mode of
a given list of numbers.

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]


```python
import statistics as stats
import pandas as pd

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

statistical_mean = stats.mean(numbers)
print(statistical_mean)

statistical_median = stats.median(numbers)
print(statistical_median)

mode_result = pd.Series(numbers).mode()
print("All modes:", mode_result.tolist())

```

Question 6: Compute the covariance and correlation coefficient between the following
two datasets provided as lists in Python:

list_x = [10, 20, 30, 40, 50]

list_y = [15, 25, 35, 45, 60]


```python
import pandas as pd
import numpy as np
data = {
    "list_x" : [10, 20, 30, 40, 50],
    "list_y" : [15, 25, 35, 45, 60]
}

 
df = pd.DataFrame(data)
df
```

```python
# calcualte covariance
cov_matrix = df.cov()# by default computes sample variance
print("Covariance Matrix : \n" , cov_matrix)
```

```python
# correlation
corr_matrix = df.corr()
print("Correlation Matrix :\n" , corr_matrix)
```

```python
"""Question 7: Write a Python script to draw a boxplot for the following numeric list and
identify its outliers. Explain the result:
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]"""

import matplotlib.pyplot as plt
import numpy as np

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Draw boxplot
plt.boxplot(data)
plt.title("Boxplot of Data")
plt.ylabel("Values")
plt.show()

# Calculate quartiles and IQR
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

# Calculate lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Outliers:", outliers)
```

```python
"""Question 8: You are working as a data analyst in an e-commerce company. The
marketing team wants to know if there is a relationship between advertising spend and
daily sales.
● Explain how you would use covariance and correlation to explore this
relationship.
● Write Python code to compute the correlation between the two lists:
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]
"""

import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

corr = np.corrcoef(advertising_spend, daily_sales)[0, 1]
print(corr)


""" 
- To check the relationship between advertising spend and daily sales, we 

use covariance and correlation.


- Covariance tells us the direction of the relationship. A positive covariance means 

that when advertising spend increases, sales also increase. A negative covariance

means that when advertising spend increases, sales decrease.

- Correlation tells us how strong the relationship is. Its values range from −1 to +1. 

A value close to +1 indicates a strong positive relationship, while a value close to 
0 indicates a weak or no relationship.


- In this case, we expect a positive relationship because higher advertising spend
usually leads to higher sales."""
```
